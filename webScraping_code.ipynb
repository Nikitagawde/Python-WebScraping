## 1) Web Scraping 

```python
from requests import get

#save the HTML of the site within the url variable
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_system_of_government'

response = get(url)
response
```

```python
#import library to parse HTML from page
from bs4 import BeautifulSoup as bs

#parse data from "page" and save to new variable "soup"
soup = bs(response.text,'html.parser')

#displaying the type of variable "soup"
type(soup)
```

```python
#pinpointing the location of the table and its contents
my_table = soup.find('table', attrs= {'class': 'wikitable sortable'}).tbody
my_table
```

```python
#list to store scraped value data in:


countries = []
constitution = []
head_of_state = []


#utilizing HTML tags for rows <tr> and elements <td> to iterate through each row of 
#data and append data elements to their appropriate lists
for rows in my_table.find_all('tr'):
    cells = rows.find_all('td')
    if len(cells) == 4:
        countries.append(cells[0].find('a').text)
        constitution.append(cells[1].find(text = True).replace('\n', '').replace('\xa0',''))
        head_of_state.append(cells[2].find(text = True).replace('\n', ''))
        
constitution
```

```python
#create the table data from the html page in pandas dataframe
test_df = pd.DataFrame({"Country" : countries,
                        "constitution" : constitution,
                        "head_of_state":head_of_state
    
})
```

```python
#cleaning dataframe of non-Ascii characters

test_df.Country.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)

#display the dataframe
test_df
```
